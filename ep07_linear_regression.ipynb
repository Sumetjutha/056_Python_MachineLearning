```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,100)
```

```python
x
```

```python
y = (2*x) + 1
```

```python
y
```

```python
plt.plot(x,y,'-r',label='y=2x+1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc="upper center")
plt.title("Graph y = 2x + 1")
plt.grid()
```
